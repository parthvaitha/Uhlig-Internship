```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
path_to_data_file = './Map-Q12021_data.xlsx'
df1 = pd.read_excel(path_to_data_file)
df1.rename(columns={"Full County Number": "FIPS"}, inplace = True)
df1.set_index('FIPS', inplace = True)
df1
```

```python
path_to_data_file2 = './RuralAtlasData23.xlsx'
df2 = pd.read_excel(path_to_data_file2, 'People')
df2.set_index('FIPS', inplace = True)
df2
```

```python
df3 = pd.read_excel(path_to_data_file2, 'Jobs')
df4 = pd.read_excel(path_to_data_file2, 'County Classifications')
df5 = pd.read_excel(path_to_data_file2, 'Veterans')
df6 = pd.read_excel(path_to_data_file2, 'Income')
```

```python
df69 = df2.merge(df3,how='left')
df69 = df69.merge(df4, how='left')
df69 = df69.merge(df5, how='left')
df69 = df69.merge(df6, how='left')
# df69 = df69.merge(df1, how = 'left')
df69
```

```python
df69.set_index('FIPS', inplace = True)
df69
```

```python
df69 = df69.merge(df1, on = ['FIPS'])
```

```python
df69
```

```python
blackpopchangerate = []
immigrationrate00_10 = []
hispanicpct2010 = []
popdensity2010 = []
asianpct2010 = []
blackpct2010 = []
nativepct2010 = []
white2010 = []
alphax = []
adjr = []
```

```python
for i in ['UnempRate2020', 'PerCapitaInc', 'Poverty_Rate_ACS', 'MedHHInc', 'Q1 2021']:
    y = df69[i]
    X = sm.add_constant(df69[['NonHispanicBlackPopChangeRate0010',
                             'Immigration_Rate_2000_2010', 
                             'HispanicPct2010', 
                             'PopDensity2010',
                             'AsianNonHispanicPct2010',
                             'BlackNonHispanicPct2010',
                             'NativeAmericanNonHispanicPct2010',
                             'WhiteNonHispanicPct2010']])
    results = sm.OLS(y,X,missing='drop').fit()
    alpha = (results.params['const'])
    alphax.append(alpha)
    blackpopchangerate.append(results.params['NonHispanicBlackPopChangeRate0010'])
    immigrationrate00_10.append(results.params['Immigration_Rate_2000_2010'])
    hispanicpct2010.append(results.params['HispanicPct2010'])
    popdensity2010.append(results.params['PopDensity2010'])
    asianpct2010.append(results.params['AsianNonHispanicPct2010'])
    blackpct2010.append(results.params['BlackNonHispanicPct2010'])
    nativepct2010.append(results.params['NativeAmericanNonHispanicPct2010'])
    white2010.append(results.params['WhiteNonHispanicPct2010'])
    adjr.append(results.rsquared_adj)
```

```python
ad = pd.DataFrame(df69['UnempRate2020'].describe())
bd = pd.DataFrame(df69['PerCapitaInc'].describe())
cd = pd.DataFrame(df69['Poverty_Rate_ACS'].describe())
dd = pd.DataFrame(df69['MedHHInc'].describe())
ed = pd.DataFrame(df69['Q1 2021'].describe())
test3 = pd.concat([ad, bd, cd, dd, ed], axis=1)
test3.columns = ['Unemployment Rate 2020', 
                 'Per Capita Income (2019 inflation adjusted)', 
                 'Poverty Rate (2015-19)', 
                 'Median Houshold Income (2019 inflation adjusted)',
                 'Property Values Q1 2021']
test3
```

```python
finale = pd.DataFrame({'Dependent Variable':['Unemployment Rate 2020', 
                                             'Per Capita Income', 
                                             'Poverty Rate', 
                                             'Median Household Income', 
                                             'Property Values Q12021'],
                       'Alpha':alphax,
                       'Black Population Change Rate (2000-2010)':blackpopchangerate,
                       'Immigration Rate (2000-2010)':immigrationrate00_10,
                       'Percentage Hispanic 2010':hispanicpct2010,
                       'Percentage Asian 2010':asianpct2010,
                       'Percentage Black 2010': blackpct2010,
                       'Percentage Native American': nativepct2010,
                       'Percentage White 2010': white2010,
                       'Population Density 2010': popdensity2010,
                       'Adjusted R Squared': adjr})
finale.set_index('Dependent Variable', inplace = True)
finale
```
